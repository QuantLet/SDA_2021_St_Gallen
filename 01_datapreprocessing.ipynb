```python
#packages
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import pandas as pd
from io import StringIO
```

# Data Preprocessing

## Cryto Data

```python
#automated google chrome webbrowser
driver = webdriver.Chrome('C:/Users/mario/Desktop/MBF_St.Gallen/3. Semester/Smart Data Analytics/chromedriver/chromedriver')
driver.get('https://coinmarketcap.com/currencies/bitcoin/historical-data/')
```

```python
#historical data
data = driver.find_element_by_xpath("//*[@id='__next']/div[1]/div/div[2]/div/div[3]/div/div/div[2]/table").text
```

```python
#data preprocessing
datastr = data.replace('Dec ', '12/').replace('Nov ', '11/').replace('Oct ', '10/').replace('Sep ', '09/').replace('Aug ', '08/').replace('Jul ', '07/').replace('Jun ', '06/').replace('May ', '05/').replace('Apr ', '04/').replace('Mar ', '03/').replace('Feb ', '02/').replace('Jan ', '01/').replace(', 2021', '/2021').replace(', 2020', '/2020').replace(', 2019', '/2019').replace(', 2018', '/2018').replace(', 2017', '/2017').replace('$','').replace(',','')
datastr = StringIO(datastr)
df = pd.read_csv(datastr, sep=" ",index_col=False)
df.drop('Cap', inplace=True, axis=1)
df = df.rename(columns={'Market':'Market Cap'})
```

```python
df
```

```python
df['Date'] = pd.to_datetime(df['Date'],
                                format="%m/%d/%Y",
                                errors='raise')
df_btc_price = df
```

```python
driver.get('https://coinmarketcap.com/currencies/ethereum/historical-data/')
```

```python
#historical data
data = driver.find_element_by_xpath("//*[@id='__next']/div[1]/div/div[2]/div/div[3]/div/div/div[2]/table").text
```

```python
#data preprocessing
datastr = data.replace('Dec ', '12/').replace('Nov ', '11/').replace('Oct ', '10/').replace('Sep ', '09/').replace('Aug ', '08/').replace('Jul ', '07/').replace('Jun ', '06/').replace('May ', '05/').replace('Apr ', '04/').replace('Mar ', '03/').replace('Feb ', '02/').replace('Jan ', '01/').replace(', 2021', '/2021').replace(', 2020', '/2020').replace(', 2019', '/2019').replace(', 2018', '/2018').replace(', 2017', '/2017').replace('$','').replace(',','')
datastr = StringIO(datastr)
df = pd.read_csv(datastr, sep=" ",index_col=False)
df.drop('Cap', inplace=True, axis=1)
df = df.rename(columns={'Market':'Market Cap'})

df['Date'] = pd.to_datetime(df['Date'],
                                format="%m/%d/%Y",
                                errors='raise')
df_eth_price = df
```
